```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/pacs_10_runs.csv')
```

```python
df.keys()
```

```python
data = df.to_numpy()
data.shape
```

```python
X = data[:,0]
Y = data[:,1::3]
```

```python
X.shape, Y.shape
```

```python
Y1 = Y[:,:10]; M1 = Y1.mean(axis=1); S1 = Y1.std(axis=1)
Y2 = Y[:,10:]; M2 = Y2.mean(axis=1); S2 = Y2.std(axis=1)
```

```python
plt.plot(X, Y[:,:10].mean(axis=1), 'b', label='BT + MixStyle')
plt.fill_between(X, M1+S1, M1-S1, facecolor='b', alpha=0.5)


plt.plot(X, Y[:,10:].mean(axis=1), 'r', label='BT')
plt.fill_between(X, M2+S2, M2-S2, facecolor='r', alpha=0.5)

plt.legend(loc='best')
```
